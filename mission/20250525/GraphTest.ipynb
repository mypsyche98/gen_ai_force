```python
# 구름이의 그래프 탐구 https://level.goorm.io/exam/351299/%EA%B5%AC%EB%A6%84%EC%9D%B4%EC%9D%98-%EA%B7%B8%EB%9E%98%ED%94%84-%ED%83%90%EA%B5%AC/quiz/1

class Graph:
  def __init__(self, nodeIdx):
    self.nodeIdx = nodeIdx
    self.edgeList = []

  def get_idx(self):
    return self.nodeIdx

  def get_edgeList(self):
    return self.edgeList

  def add_edge(self, edgeIdx):
    if edgeIdx not in self.edgeList:
      self.edgeList.append(edgeIdx)

  def __repr__(self):
    return f"nodeIdx={self.nodeIdx}/edgeCnt={len(self.edgeList)}"

  def get_edge_count(self):
    return len(self.edgeList)

  def traverse(self, graph_all):
    totalNodeList = []
    totalNodeList.append(self.get_idx())

    candiNodeList = self.edgeList.copy()
    print(f"total={totalNodeList} / candi={candiNodeList}")


    #while len(candiNodeList) > 0:
    for xx in candiNodeList:
      node = graph_all[xx]
      if node.get_idx() not in totalNodeList:
        totalNodeList.append(node.get_idx())

      for x in node.get_edgeList():
        if x not in totalNodeList:
          candiNodeList.append(x)

    return len(totalNodeList)

graph_all = {}


#data_count = int(input())
#for i in range(data_count):
data = """1 2
2 3
3 4
5 6
7 8
9 10"""
for str in data.split("\n"):
  #print(f"[{str}]")

  nodeIdx, edgeIdx = str.split()

  if int(nodeIdx) == int(edgeIdx):
    continue

  if nodeIdx not in graph_all:
    graph_all[nodeIdx] = Graph(nodeIdx)

  graph_all[nodeIdx].add_edge(edgeIdx)

  if edgeIdx not in graph_all:
    graph_all[edgeIdx] = Graph(edgeIdx)

  graph_all[edgeIdx].add_edge(nodeIdx)

#print(graph_all)
#print(graph_all["1"].traverse(graph_all))

edge_cnt = []
for graph in graph_all.values():
  friends_count = graph.traverse(graph_all)
  if friends_count > 1:
    edge_cnt.append(friends_count)

print(edge_cnt)

if len(edge_cnt) > 0:
  edge_cnt = sorted(edge_cnt)
  print(edge_cnt[0], edge_cnt.pop())
else :
	print("0 0")
```
