```python
# queue.py

import sys

class Queue:
  def __init__(self, *args):
    if len(args) > 0:
      self.size = int(args[0])
    else:
      self.size = 0
    self.queue = []

  def put(self, data):
    if self.size == 0 or len(self.queue) < self.size:
      self.queue.append(data)
    else:
      print("Overflow")

  def get(self):
    if len(self.queue) == 0:
      print("Underflow")
    else:
      return self.queue.pop(0)

  def get_size(self):
    if self.size == 0:
      return sys.maxsize
    else:
      return self.size

```

```python
# size가 없는 queue

queue_of_inf = Queue()
print("size=%d" % queue_of_inf.get_size())

queue_of_inf.get()
queue_of_inf.put("A")
queue_of_inf.put(1)
print(queue_of_inf.get())
print(queue_of_inf.get())
print(queue_of_inf.get())
```

```python
# size가 있는 queue

queue_of_limit = Queue(5)

print("size=%d" % queue_of_limit.get_size())

queue_of_limit.get()
queue_of_limit.put("A")
queue_of_limit.put(1)
print(queue_of_limit.get())
print(queue_of_limit.get())
print(queue_of_limit.get())
queue_of_limit.put(1)
queue_of_limit.put(2)
queue_of_limit.put(3)
queue_of_limit.put(4)
queue_of_limit.put(5)
queue_of_limit.put(6)
```
