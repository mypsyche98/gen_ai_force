```python
# 정규표현식으로 이메일 주소 추출해보기
# https://www.notion.so/6-1fc1ebda1846803da5a3e2b1760618d6?source=copy_link#2151ebda1846801b98afdf95a0b38810

"""
이메일주소는 다음과 같은 형태이다

id@domain

id - 는 보통 영대문자,영소문자와 일부특수기호가 포함된다. 가능한 특수기호는 많으나 보통 - _ . 정도만 허용된다 => 하지만 연속된 .은 허용하지 않는다.
@ - 이 문자는 반드시 한개만 존재한다
도메인 - 보통 영대소문자, 숫자, 하이픈만 가능하며 . 은 여러개 올 수 있으나 연속된 .은 허용하지 않는다.
"""

import re

p = re.compile("^[a-zA-Z0-9-.+_]+@[\w-]+\.[a-zA-Z0-9-.]+$")

emails = ['python@mail.example.com', 'python+kr@example.com',              # 올바른 형식
          'python-dojang@example.co.kr', 'python_10@example.info',         # 올바른 형식
          'python.dojang@e-xample.com',                                    # 올바른 형식
          '@example.com', 'python@example', 'python@example-com']          # 잘못된 형식

for email in emails:
  print("email=",p.match(email) != None)
```

```python
# 벡터의 기본 연산 수행하기 - 덧셈, 뺄셈, 곱셈

import numpy as np

# 벡터 생성
u = np.array([2, 3])
v = np.array([1, -1])

# 덧셈
sum_result = u + v
print("덧셈 결과:", sum_result)  # 출력 예: [3 2]

# 뺄셈
sub_result = u - v
print("뺄셈 결과:", sub_result)  # 출력 예: [1 4]

# 스칼라 곱
scalar = 2
scalar_result = scalar * u
print("스칼라 곱 결과:", scalar_result)  # 출력 예: [4 6]

```

```python
# 선형연립방정식
#(1) np.linalg.solve 를 활용한 방법
import numpy as np
A = np.array([[2,3],[5,6]])
B = np.array([4, 5])
C = np.linalg.solve(A, B)
print(C)
```

```python
#(2) np.linalg.inv (역행렬 활용)
import numpy as np
A = np.array([[2,3],[5,6]])
B = np.array([4, 5])
C = np.linalg.inv(A)
D = np.dot(C, B)
print(D)
```
