```python
# https://level.goorm.io/exam/51358/%EC%9E%AC%EC%9B%90-%EB%84%98%EB%B2%84/quiz/1
# 재원넘버

import sys
#input = sys.stdin.readline

N = int(input().rstrip())

# 재원넘버는 2,3,6 이므로 각 자리에 2,3,6은 한개씩 밖에 없음
# 따라서 자리수별로 3**n 을 하여 더하기함

sum = 0
for x in range(1,N+1):
  sum += 3 ** x

print(sum)

# 제출결과 PASS 10/10
```

```python
# https://level.goorm.io/exam/195693/%ED%86%B5%EC%A6%9D-2/quiz/1
# 통증(2)

import sys
#input = sys.stdin.readline

pain = int(input().rstrip())
potionA, potionB = map(int, input().rstrip().split())

min_potion = min(potionA, potionB)
max_potion = max(potionA, potionB)

if pain < min_potion:
  print(-1)
elif min_potion < pain < max_potion:
  if pain % min_potion > 0:
    print(-1)
  else:
    print(pain // min_potion)
else:
  max_n = pain // max_potion

  while max_n > 0 and (pain-(max_potion*max_n)) % min_potion != 0:
    #print("max_n:", max_n)
    max_n -= 1

  if (pain-(max_potion*max_n)) % min_potion == 0:
    print(max_n+((pain-(max_potion*max_n))//min_potion))
  else:
    print(-1)

# 제출결과 PASS 26/26
```

```python
# https://level.goorm.io/exam/49078/%EB%8F%99%EC%A0%84-%EC%A4%8D%EA%B8%B0-%EB%8C%80%ED%9A%8C/quiz/1
# 동전 줍기 대회

from itertools import accumulate
import sys
input = sys.stdin.readline

num_of_coins = int(input().rstrip())
coins = list(map(int, input().rstrip().split())) + [0]

prefix_sum = list(accumulate(coins))

# 누적합 중 MAX
max_money = max(prefix_sum)
curr_max = 0

for x in range(num_of_coins):
  curr_max = max(curr_max + coins[x], coins[x])
  if max_money < curr_max :
    max_money = curr_max

print(max_money)

# 제출결과 PASS 10/10
```
