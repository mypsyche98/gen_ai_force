```python
# https://level.goorm.io/exam/353492/%EB%8D%94%ED%95%98%EA%B8%B0%EC%99%80-%EA%B3%B1%ED%95%98%EA%B8%B0/quiz/1
# 더하기와 곱하기

# 이 문제는 에시 2번의 답이 아래 설명처럼 오답으로 판단되어 문의를 남겼습니다

import sys
input = sys.stdin.readline

N, K = map(int, input().rstrip().split())
arr = list(map(int, input().rstrip().split()))

mod = 10**9 + 7
dp = [{} for x in range(N)]

dp[0][0] = 0
dp[0][arr[0]] = 1

for x in range(1,N):
    tgt = arr[x]

    for key, val in dp[x-1].items():
        added = (key + tgt) % 1000

        dp[x][added] = (dp[x].get(added, 0) + val) % mod

        muled = (key * tgt) % 1000

        dp[x][muled] = (dp[x].get(muled, 0) + val) % mod

print(dp[-1][K] % mod)

# 제출결과 PASS 23/23
```
