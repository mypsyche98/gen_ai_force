```python
# https://www.acmicpc.net/problem/1992
# 쿼드트리

import sys
#input = sys.stdin.readline

N = int(input().rstrip())

matrix = []
for x in range(N):
  matrix.append(input())

def quad_tree(submatrix, n, y, x):
  #print(n,x,y)
  for i in range(y, y+n):
    for j in range(x, x+n):
      if submatrix[y][x] != submatrix[i][j]:
        half_n = n // 2
        result = '('
        result += quad_tree(submatrix, half_n, y, x)
        result += quad_tree(submatrix, half_n, y, x + half_n)
        result += quad_tree(submatrix, half_n, y + half_n, x)
        result += quad_tree(submatrix, half_n, y + half_n, x + half_n)
        result += ')'
        return result

  return submatrix[y][x]

print(quad_tree(matrix, N, 0, 0))

```

```python
# https://www.acmicpc.net/problem/1074
# Z

import sys
#input = sys.stdin.readline

N, R, C = map(int, input().rstrip().split())

answer = 0

def find_pos(n, r, c, R, C):
  global answer
  if r == R and c == C:
    print(answer)
    return

  if not(r <= R < r+n and c <= C < c+n):
    answer += n * n
    return

  half_n = n // 2
  find_pos(half_n, r, c, R, C)
  find_pos(half_n, r, c+half_n, R, C)
  find_pos(half_n, r+half_n, c, R, C)
  find_pos(half_n, r+half_n, c+half_n, R, C)

find_pos(2**N, 0, 0, R, C)

```

```python
# https://www.acmicpc.net/problem/4779
# 칸토어 집합

def solve(n):
  if n == 1:
    return "-"

  left = solve(n // 3)
  center = ' ' * (n // 3)
  return left + center + left

while True:
    try:
        N = int(input())

        rst = solve(3 ** N)
        print(rst)
    except:
        break

```
