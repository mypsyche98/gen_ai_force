```python
# https://level.goorm.io/exam/348461/%ED%9A%A8%EC%9C%A8%EC%A0%81%EC%9D%B8-%EC%A7%91%EC%95%88%EC%9D%BC/quiz/1
# 효율적인집안일

import sys
input = sys.stdin.readline

T = int(input().rstrip())
C = int(input().rstrip())
M = []

for x in range(C):
  M.append(int(input().rstrip()))


#시간이 적게 걸리는 작업부터 해야 최대한 많이 할 수 있다.
M.sort()
#print(M)

cnt = 0
for w in M:
  if cnt == C:
    break

  if T > w:
    T -= w
    cnt += 1
  else:
    break

print(cnt)

# 제출결과 FAIL 9/10
```

```python
# https://level.goorm.io/exam/195697/%EA%B3%BC%EC%9D%BC-%EA%B5%AC%EB%A7%A4/quiz/1
# 과일구매

import sys
#input = sys.stdin.readline

num_of_fruits, sum_of_money = map(int, input().rstrip().split())

# P - 과일가격
# C - 포만감
# X - C/P 가격1당 포만감
fruits = []
for x in range(num_of_fruits):
  P, C = list(map(int, input().rstrip().split()))
  fruits.append([P, C, C//P])

# 조각포만감이 큰순
# 가격이 적은순
# 전체포만감이 큰순
fruits.sort(key=lambda x : (-x[2], x[0], -x[1]))
#print(fruits)

sum_of_C = 0
for fruit in fruits:
  if sum_of_money >= fruit[0]:
    sum_of_C += fruit[1]    #포만감
    sum_of_money -= fruit[0]  #돈
  else:
    sum_of_C += (fruit[2] * sum_of_money) # 나머지돈은 모두 조각과일
    break

print(sum_of_C)

# 재출결과 PASS 26/26
```

```python
# https://level.goorm.io/exam/175194/%EA%B5%AC%EB%A6%84-%EC%8A%A4%ED%80%98%EC%96%B4/quiz/1
# 구름스퀘어

import sys
input = sys.stdin.readline

num_of_events = int(input().rstrip())

# [0] = 시작시간
# [1] = 종료시간
event_list = []
for x in range(num_of_events):
	event_list.append(list(map(int, input().rstrip().split())))

# 종료시간이 빠른순
# 시작시간이 빠른순
event_list.sort(key=lambda x : (x[1], x[0]))

#print(event_list)

count = 0
before_end_time = 0

for start, end in event_list:
	# 이전 회의 1시간뒤부터 시작가능하므로 무조건 커야한다.
	if start > before_end_time:
		count += 1
		before_end_time = end

print(count)

# 제출결과 PASS 15/15
```

```python
# https://level.goorm.io/exam/163680/%EC%A7%81%EC%82%AC%EA%B0%81%ED%98%95-%EB%A7%8C%EB%93%A4%EA%B8%B0/quiz/1
# 사각형만들기

import sys
input = sys.stdin.readline

num_of_sticks = int(input().rstrip())

stick_dict = {}
for x in map(int, input().rstrip().split()):
	if x in stick_dict:
		stick_dict[x] += 1
	else:
		stick_dict[x] = 1

# 두개씩 쌍을 만듦
stick_pair = []
for length, cnt in stick_dict.items():
	while cnt > 1:
		stick_pair.append(length)
		cnt -= 2

stick_pair.sort(reverse=True)
total_area = 0
for x in range(1, len(stick_pair), 2):
	total_area += stick_pair[x-1] * stick_pair[x]


print(total_area)

# 제출결과 PASS 40/40
```
