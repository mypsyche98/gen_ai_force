```python
# 두 수위 최대공약수 구하기 : GCD: Greatest Common Divisor

# 유클리드 호제법 - 두 수 중 큰 수를 작은 수로 나누어 나머지를 구하고, 다시 작은 수를 나머지로 나누는 과정을 반복하여 나머지가 0이 되었을 때의 나누는 수가 최대공약수이다

def getGCD(num_a, num_b):
  if num_a == num_b :
    return num_a

  bigger = max(num_a, num_b)
  smaller = min(num_a, num_b)

  while bigger % smaller:
    bigger, smaller = smaller, bigger % smaller

  return smaller

num_as = [1071, 24, 45, 13, 10]
num_bs = [1029, 36, 75, 52, 10]
answers = [21, 12, 15, 13, 10]

for num_a, num_b, answer in zip(num_as, num_bs, answers):
  print(f"{num_a} {num_b}의 GCD는 {getGCD(num_a,num_b)} 이다 = 정답({answer})")
```

```python
# 최소공배수 구하기 LCM: Least Common Multiple
# 최대공약수를 이용하는 방법 - 두수의 최소공배수는 두수의 곱을 최대공약수로 나눈것과 같다

def getLCM(num_a, num_b):
  return (num_a*num_b)/getGCD(num_a, num_b)

num_as = [24, 18, 12, 36, 12]
num_bs = [40, 25, 18, 45, 15]
answers = [120, 450, 36, 180, 60]

for num_a, num_b, answer in zip(num_as, num_bs, answers):
  print(f"{num_a} {num_b}의 LCM는 {getLCM(num_a,num_b)} 이다 = 정답({answer})")

```

```python
# 큐를 이용한 프린터 작업 스케줄링 시뮬레이션

# 우선순위 큐를 이용한 프린터 스플링 시뮬레이션 => 작업에 우선순위 값을 넣어 해당 데이터가 먼저 출력 되도록 한다

class print_spool():
  def __init__(self, max_size=5):
    self.heap = []
    self.max = max_size

  def add_job(self, text, priority):
    if len(self.heap) == self.max:
      return "SPOOL IS FULL!!!"

    heapq.heappush(self.heap, (-priority, text))

  def print_job(self):
    if self.heap:
      job = heapq.heappop(self.heap)
      print(f"PRINT - {job[1]}")
    else:
      print("SPOOL IS EMPTY")

  def remove_job(self, text):
    for x in range(len(self.heap)):
      if self.heap[x][1] == text:
        print("SPOOL REMOVED ["+text+"]")
        del self.heap[x]
        return

printer = print_spool()
printer.print_job()
printer.add_job("1st-5", 5)
printer.add_job("2nd-10", 10)
printer.print_job()
printer.print_job()
print()
printer.add_job("1st-5", 5)
printer.add_job("2nd-10", 10)
printer.add_job("3rd-6", 6)
printer.print_job()
printer.add_job("4th-20", 20)
printer.print_job()
printer.remove_job("3rd-6")s
printer.print_job()
```
