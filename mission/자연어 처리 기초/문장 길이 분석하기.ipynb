```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns # 시각화를 위해 seaborn 사용
```

```python
sample_texts = [
  "This is a short sentence.",
  "This one is a little bit longer.",
  "NLP missions are always fun and interesting!",
  "Text length analysis is a fundamental step in EDA.",
  "Wow!",
  "Analyzing the distribution of sentence lengths can reveal characteristics of the dataset.",
  "Short texts, like tweets, will have a very different distribution compared to news articles.",
  "We are learning NLP.",
  "This is step two."
]
```

```python
df = pd.DataFrame({'text': sample_texts})
```

```python
# 3-1. 글자 수 (공백 포함)
df['char_length'] = df['text'].apply(len)
```

```python
# 3-2. 단어 수 (공백으로 쪼개기)
# (이것이 더 중요한 지표입니다)
df['word_count'] = df['text'].apply(lambda x: len(x.split()))
```

```python

print("--- 1. 문장 길이 계산 결과 ---")
print(df.head())
```

```python
# 4. 문장 길이 통계 요약
print("\n--- 2. 단어 수(word_count) 기술 통계 ---")
print(df['word_count'].describe())
```

```python

# 5. 문장 길이 분포 시각화 (히스토그램)
print("\n--- 3. 단어 수 분포 시각화 (Histogram) ---")

plt.figure(figsize=(10, 5))
# displot (KDE 포함) 또는 histplot 사용
sns.histplot(df['word_count'], bins=10, kde=True)

plt.title('Distribution of Sentence Length (Word Count)')
plt.xlabel('Word Count')
plt.ylabel('Frequency')
plt.show() #
```
