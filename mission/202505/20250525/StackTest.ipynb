```python
# stack.py

import sys

class Stack:
  def __init__(self, *args):
    if len(args) > 0:
      self.size = int(args[0])
    else:
      self.size = 0
    self.stack = []

  def push(self, data):
    if self.size == 0 or len(self.stack) < self.size:
      self.stack.append(data)
    else:
      print("Overflow")

  def pop(self):
    if len(self.stack) == 0:
      print("Underflow")
    else:
      return self.stack.pop()

  def get_size(self):
    if self.size == 0:
      return sys.maxsize
    else:
      return self.size

```

```python
# size가 없는 stack

stack_of_inf = Stack()
print("size=%d" % stack_of_inf.get_size())

stack_of_inf.pop()
stack_of_inf.push("A")
stack_of_inf.push(1)
print(stack_of_inf.pop())
print(stack_of_inf.pop())
print(stack_of_inf.pop())
```

```python
# size가 있는 stack

stack_of_limit = Stack(5)

print("size=%d" % stack_of_limit.get_size())

stack_of_limit.pop()
stack_of_limit.push("A")
stack_of_limit.push(1)
print(stack_of_limit.pop())
print(stack_of_limit.pop())
print(stack_of_limit.pop())
stack_of_limit.push(1)
stack_of_limit.push(2)
stack_of_limit.push(3)
stack_of_limit.push(4)
stack_of_limit.push(5)
stack_of_limit.push(6)
print(stack_of_limit.pop())
```
