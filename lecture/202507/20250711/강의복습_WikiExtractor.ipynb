```python
import zipfile
import os

# 'extracted_files' 디렉토리가 없으면 생성
output_dir = 'kowiki'
os.makedirs(output_dir, exist_ok=True)

with zipfile.ZipFile('kowiki-latest-pages-articles.zip', 'r') as zf:
    zf.extractall(output_dir) # 지정된 디렉토리에 모든 파일 압축 해제
    print(f"\n모든 파일이 '{output_dir}' 디렉토리에 압축 해제되었습니다.")

# 압축 해제된 파일 목록 확인 (예시)
print(f"'{output_dir}' 디렉토리 내용:", os.listdir(output_dir))
```
