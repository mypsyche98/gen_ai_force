```python
import numpy as np
from numpy import dot
from numpy.linalg import norm

def cos_sim(A, B):
  return dot(A, B)/(norm(A)*norm(B))

vec1 = np.array([0,1,1,1])
vec2 = np.array([1,0,1,1])
vec3 = np.array([2,0,2,2])

print('벡터1과 벡터2의 유사도 :',cos_sim(vec1, vec2))
print('벡터1과 벡터3의 유사도 :',cos_sim(vec1, vec3))
print('벡터2와 벡터3의 유사도 :',cos_sim(vec2, vec3))
```

```python
data = ['저는 배가 고파요',
        '저기 배가 지나가네요',
        '굶어서 허기가 지네요',
        '허기 워기라는 게임이 있는데 즐거워',
        '스팀에서 재밌는 거 해야지',
        '스팀에어프라이어로 연어구이 해먹을거야']

df = pd.DataFrame(data, columns=['text'])
df
```

```python
def get_embedding(text):
  return embeddings.embed_query(text)

def get_embedding_new(text):
  response = client.embeddings.create(
      input=text,
      model="text-embedding-ada-002"
  )
  return response.data[0].embedding


df['embedding'] = df.apply(lambda row: get_embedding_new(
        row.text,
    ), axis=1)
df
```

```python
def cos_sim(A, B):
  return dot(A, B)/(norm(A)*norm(B))

def return_answer_candidate(df, query):
    query_embedding = get_embedding_new(query)
    df["similarity"] = df.embedding.apply(lambda x: cos_sim(np.array(x), np.array(query_embedding)))
    top_three_doc = df.sort_values("similarity", ascending=False).head(3)
    return top_three_doc
```

```python
sim_result = return_answer_candidate(df, '아무것도 안 먹었더니 꼬르륵 소리가 나네')
sim_result
```

```python
os.environ['OPENAI_API_KEY'] = get_key()

# 객체 생성
llm = ChatOpenAI(
    temperature=0.1,  # 창의성 (0.0 ~ 2.0)
    max_tokens=2048,  # 최대 토큰수
    model_name="gpt-4.1",  # 모델명
)
```

```python
# 질의내용
question = "세종대왕이 누구인지 설명해주세요"
# 질의
result = llm.invoke(question)
print(result.content)
```

```python
# 질문 템플릿 형식 정의
template = "{who}가 누구인지 설명해주세요"

# 템플릿 완성
prompt = PromptTemplate(
        template=template, input_variables=['who']
    )
print(prompt)
```

```python
print(prompt.format(who="에이다 러브레이스"))
```

```python
# 연결된 체인 생성
llm_chain = prompt | llm
result = llm_chain.invoke({"who":"리처트 파인만"})
print(result.content)
```

```python
result = llm_chain.invoke({"who":"베르나르 베르베르"},
                          config={"callbacks": [ConsoleCallbackHandler()]})
```

```python
# 프롬프트 템플릿 생성
template = """아래는 사람과 AI의 친근한 대화입니다. AI의 이름은 프시케봇입니다. 대화 문맥을 바탕으로 친절한 답변을 진행하세요.

Current Conversation:
{history}

Human: {input}
AI:"""

prompt = PromptTemplate(
        template=template, input_variables=['history', 'input']
    )

# llm 객체를 새로 선언하고, 프롬프트 템플릿과 llm 객체를 연결합니다.
llm = ChatOpenAI(model_name="gpt-4.1")
chain = prompt | llm
```

```python
store = {}
session_id = "session20250723"
```

```python
if session_id not in store:
    store[session_id] = ChatMessageHistory()
session_history = store[session_id]
```

```python
with_message_history = RunnableWithMessageHistory(
    chain,
    lambda session_id: session_history,
    input_messages_key="input",
    history_messages_key="history"
)
```

```python
result = with_message_history.invoke(
    {"input": "당신은 어디에서 만들었습니까?"},
    config={"configurable": {"session_id": session_id}},
)
print(result.content)
```

```python
result = with_message_history.invoke(
    {"input": "푸른 바다를 주제로 감성적이고 짧은 시를 하나 지어주세요"},
    config={"configurable": {"session_id": session_id}},
)
print(result.content)
```

```python
result = with_message_history.invoke(
    {"input": "석양을 주제로도 해줘"},
    config={"configurable": {"session_id": session_id}},
)
print(result.content)
```
