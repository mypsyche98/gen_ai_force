```python
# page76

import sklearn
import numpy as np
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score

iris = load_iris()

X_train, X_test, y_train, y_test = train_test_split(iris.data, iris.target, test_size=0.3, random_state=1)

dt_clf = DecisionTreeClassifier(random_state=1)
dt_clf.fit(X_train, y_train)
pred = dt_clf.predict(X_test)

print("accuracy_score=", round(accuracy_score(y_test, pred), 4))
```

```python
# page77

from sklearn.model_selection import KFold
kfold = KFold(n_splits=5)

iris = load_iris()
feature = iris.data
label = iris.target

cv_accuracy = []
n_iter = 0
dt_clf = DecisionTreeClassifier(random_state=1)
```

```python
# page78

print("feature.shape=", feature.shape)
```

```python
# page79

n_iter = 0
for train_index, test_index in kfold.split(feature):
  x_train, x_test = feature[train_index], feature[test_index]
  y_train, y_test = label[train_index], label[test_index]

  dt_clf.fit(x_train, y_train)
  pred = dt_clf.predict(x_test)
  n_iter += 1

  accuracy = np.round(accuracy_score(y_test, pred), 4)
  train_size = x_train.shape[0]
  test_size = x_test.shape[0]

  print("\n#{0} 교차 검증 정확도:{1}, 학습 데이터 크기:{2}, 검증 데이터 크기:{3}".format(n_iter, accuracy, train_size, test_size))
  print("#{0} 검증 세트 인덱스:{1}".format(n_iter, test_index))
  cv_accuracy.append(accuracy)
```

```python
# page80

print("\n평균 검증 정확도 : ", np.mean(cv_accuracy))

iris_df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
iris_df['label'] = iris.target
print(iris_df.head(5))
```

```python
# page87

import sklearn
import numpy as np
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from sklearn.model_selection import cross_val_score

iris = load_iris()
feature = iris.data
label = iris.target

dt_clf = DecisionTreeClassifier(random_state=1)

basic_scores = cross_val_score(dt_clf, feature, label)

print("cross_val_score 를 통한 성능평가:", basic_scores)
```

```python
# page88

print("\n평균 검증 정확도 : ", round(np.mean(basic_scores), 4))
```

```python
# page89

from sklearn.model_selection import StratifiedKFold

S_kfold = StratifiedKFold(n_splits=5)

cv_accuracy = []
n_iter = 0

for train_index, test_index in S_kfold.split(feature, label):
  x_train, x_test = feature[train_index], feature[test_index]
  y_train, y_test = label[train_index], label[test_index]

  dt_clf.fit(x_train, y_train)
  pred = dt_clf.predict(x_test)
  n_iter += 1

  accuracy = np.round(accuracy_score(y_test, pred), 4)
  cv_accuracy.append(accuracy)

print("\n평균 검증 정확도 : ", round(np.mean(cv_accuracy), 4))
```

```python
# page90

from sklearn.model_selection import cross_validate

c_val = cross_validate(dt_clf, feature, label)

df = pd.DataFrame(c_val, index=['set1','set2','set3','set4','set5'])
print(df)
```

```python
# page93

from sklearn.model_selection import LeaveOneOut

iris = load_iris()
feature = iris.data
label = iris.target

dt_clf = DecisionTreeClassifier(random_state=1)

loocv = LeaveOneOut()
score = cross_val_score(dt_clf, feature, label, cv=loocv)

print("교차검증횟수=", len(score))
print("평균검증정확도=", round(np.mean(score), 4))
```

```python
# page95

from sklearn.model_selection import ShuffleSplit

iris = load_iris()
feature = iris.data
label = iris.target

dt_clf = DecisionTreeClassifier(random_state=1)

shuffle_split = ShuffleSplit(test_size=.6, train_size=.4, n_splits=20)
score = cross_val_score(dt_clf, feature, label, cv=shuffle_split)

print("교차검증횟수=", len(score))
print("평균검증정확도=", round(np.mean(score), 4))
```

```python
# page96

from sklearn.model_selection import RepeatedStratifiedKFold

iris = load_iris()
feature = iris.data
label = iris.target

dt_clf = DecisionTreeClassifier(random_state=1)

rskfold = RepeatedStratifiedKFold()
score = cross_val_score(dt_clf, feature, label, cv=rskfold)

print("교차검증횟수=", len(score))
print("평균검증정확도=", round(np.mean(score), 4))
```
