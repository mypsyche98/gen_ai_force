```python
# page12

def hour2min(hour):
  min = hour * 60
  return min

def hour2sec(hour):
  min = hour2min(hour)
  sec = 60 * min
  return sec

print("2 hours is ", hour2sec(2), " seconds")
```

```python
# page16,17

def list_sum(num_list):
  the_sum = 0
  for i in num_list:
    the_sum = the_sum + i
  return the_sum

print(list_sum([1,3,5,7,9]))
```

```python
# page19

def list_sum_recur(num_list):
  if len(num_list) == 0:
    return 0

  the_sum = num_list[0] + list_sum_recur(num_list[1:])
  return the_sum

print(list_sum_recur([1,3,5,7,9]))
```

```python
# page20

def factorial(n):
  if n == 0:
    return 1
  else:
    recursion_step = factorial(n-1)
    result = n * recursion_step
    return result

print(factorial(10))
```

```python
# page26

def print_hanoi(towers):
  max_len = 0
  for i in range(len(towers)):
    max_len += len(towers[i])

  tower_str = []
  for i in range(len(towers)):
    #print(max_len - len(towers[i]))
    for x in range(max_len - len(towers[i])):
      tower_str.append('.')
    for y in range(len(towers[i])):
      tower_str.append(towers[i][y])

  #print(tower_str)

  for x in range(max_len):
    print(f"{tower_str[x]}\t{tower_str[x+5]}\t{tower_str[x+10]}")
  print("-"*17)

def solve_hanoi(towers, nidx, from_idx, to_idx, via_idx):
  if nidx == 1:
    towers[to_idx] = [towers[from_idx].pop(0)] + towers[to_idx]
    print_hanoi(towers)
  else:
    solve_hanoi(towers, nidx-1, from_idx, via_idx, to_idx)
    towers[to_idx] = [towers[from_idx].pop(0)] + towers[to_idx]
    print_hanoi(towers)
    solve_hanoi(towers, nidx-1, via_idx, to_idx, from_idx)

towers = [[1,2,3,4,5],[],[]]

print_hanoi(towers)
solve_hanoi(towers, 5, 0, 2, 1)
```

```python
# page41

d = [0]*100
def fib_topdown(n):
  if n==1 or n==2:
    return 1
  if d[n] != 0:
    return d[n]

  d[n] = fib_topdown(n-1)+fib_topdown(n-2)
  return d[n]

print("fib_topdown=", fib_topdown(10))
```

```python
# page42


def fib_bottomup(n):
  n_0, n_1 = 0,1
  for _ in range(n):
    n_0,n_1 = n_1,n_0+n+1
  return n_0


print("fib_bottomup=", fib_bottomup(10))
```

```python
# Singly Linked-List

class SLinkedList:

  class Node:
    def __init__(self, v, n=None):
      self.value = v
      self.next = n

  def __init__(self):
    self.head = None

  def insertNode(self,v):
    if self.head is None:
      self.head = self.Node(v)
    else:
      self.head = self.Node(v, self.head)

  def printNode(self):
    if self.head is None:
      print("node not exist")
      return
    else:
      print("list structure", end='\t')
      link = self.head

      while link:
        print(link.value, '->', end=' ')
        link = link.next
      print()

  def deleteNode(self):
    if self.head is None:
      print("node not exist")
      return
    else:
      self.head = self.head.next

  def searchNode(self,v):
    if self.head is None:
      print("node not exist")
      return
    else:
      link = self.head
      index = 0
      while link:
        if v == link.value:
          return index
        else:
          link = link.next
          index += 1
      print("node not found")


if __name__ == "__main__":
  sl = SLinkedList()
  sl.printNode()

  sl.insertNode('1st')
  sl.insertNode('2nd')
  sl.insertNode('3rd')
  sl.printNode()

  print("search")
  print("index of 1st is : {}".format(sl.searchNode('1st')))
  print("index of 555 is : {}".format(sl.searchNode('555')))

  sl.deleteNode()
  sl.deleteNode()
  sl.printNode()

  sl.deleteNode()
  sl.deleteNode()
  sl.printNode()
```

```python
# Doubly Linked-List

class DLinkedList:
  class Node:
    def __init__(self,v,n=None,p=None):
      self.value = v
      self.next = n
      self.prev = p

  def __init__(self):
    self.head = None
    self.tail = None

  def insertNodeBefore(self, v):
    if self.head is None:
      self.head = self.Node(v)
      self.tail = self.head
    else:
      self.head.prev = self.Node(v,n=self.head)
      self.head = self.head.prev

  def insertNodeAfter(self, v):
    if self.tail is None:
      self.tail = self.Node(v)
      self.head = self.tail
    else :
      self.tail.next = self.Node(v,p=self.tail)
      self.tail = self.tail.next

  def deleteNodeBefore(self):
    if self.head is None:
      print("node not exist")
      return
    else:
      self.head = self.head.next
      self.head.prev = None

  def printNodeBefore(self):
    if self.head is None:
      print("node not exist")
      return
    else:
      print("list structure", end="\t")
      link = self.head

      while link:
        print(link.value, '<->', end=' ')
        link = link.next
      print()

  def printNodeAfter(self):
    if self.head is None:
      print("node not exist")
      return
    else:
      print("list structure", end="\t")
      link = self.tail

      while link:
        print(link.value, '<->', end=' ')
        link = link.prev
      print()

  def deleteNodeBefore(self):
    if self.head is None:
      print("node not exist")
      return
    else:
      self.head = self.head.next
      self.head.prev = None

  def deleteNodeAfter(self):
    if self.tail is None:
      print("node not exist")
      return
    else:
      self.tail = self.tail.prev
      self.tail.next = None

  def searchNodeBefore(self,v):
    if self.head is None:
      print("node not exist")
      return
    else:
      link = self.head
      index = 0
      while link:
        if v == link.value:
          return index
        else:
          link = link.next
          index += 1

  def searchNodeAfter(self,v):
    if self.tail is None:
      print("node not exist")
      return
    else:
      link = self.tail
      index = 0
      while link:
        if v == link.value:
          return index
        else:
          link = link.prev
          index += 1

if __name__ == '__main__':
  dl = DLinkedList()
  dl.insertNodeBefore("1stF")
  dl.insertNodeAfter("1stB")
  dl.insertNodeBefore("2ndF")
  dl.insertNodeAfter("2ndB")
  dl.printNodeBefore()
  dl.printNodeAfter()

  print("search from head")
  print("index of 2ndF is {}".format(dl.searchNodeBefore("2ndF")))
  print("index of 1stB is {}".format(dl.searchNodeBefore("1stB")))

  print("search from tail")
  print("index of 2ndB is {}".format(dl.searchNodeAfter("2ndB")))
  print("index of 1stF is {}".format(dl.searchNodeAfter("1stF")))

  dl.deleteNodeBefore()
  dl.printNodeAfter()
  dl.deleteNodeAfter()
  dl.printNodeAfter()
```
