```python
# page19 - Largest subarray with 0 sum : 이중반복문

def maxlen(arr):
  n = len(arr)
  answer = 0
  for i in range(n):
    total = 0
    left = i # 교재에는 1 로 나와있으나 그렇게되면 변형된 두번쨰 문제애서 잘못된 답 6이 나옵니다
    right = i
    for j in range(i, n):
      total += arr[j]
      if total == 0:
        right = j
        answer = max(answer, right-left+1)
  return answer

print(maxlen([15,-2,2,-8,1,7,10,23])) # -2,2,-8,1,7
print(maxlen([-10,-2,2,-8,1,7,10,23])) # -10,-2,2,-8,1,7,10
```

```python
# page22 - Largest subarray with 0 sum : 딕셔너리

from collections import defaultdict

def maxlen_dic(arr):
  dic: dict[int, int] = {0: -1}
  ans = 0
  total = 0
  for i in range(len(arr)):
    total += arr[i]
    if total in dic:
      ans = max(ans, i - dic[total])
    else:
      dic[total] = i
  #print(dic)
  return ans


print(maxlen_dic([15,-2,2,-8,1,7,10,23])) # -2,2,-8,1,7
print(maxlen_dic([-10,-2,2,-8,1,7,10,23])) # -10,-2,2,-8,1,7,10
```

```python
# page77,78,79

def preorder_list_recur(node):
  def _preorder_list_recur(node):
    if node is None:
      return

    res.append(node.data)
    _preorder_list_recur(node.left)
    _preorder_list_recur(node.right)

  res = []
  _preorder_list_recur(node)
  return res

def inorder_list_recur(node):
  def _inorder_list_recur(node):
    if node is None:
      return

    _inorder_list_recur(node.left)
    res.append(node.data)
    _inorder_list_recur(node.right)

  res = []
  _inorder_list_recur(node)
  return res

def postorder_list_recur(node):
  def _postorder_list_recur(node):
    if node is None:
      return

    _postorder_list_recur(node.left)
    _postorder_list_recur(node.right)
    res.append(node.data)

  res = []
  _postorder_list_recur(node)
  return res

def lvlorder_list_stack(root):
  res = []

  if root is None:
    return res

  queue = [root]
  while queue:
    node = queue.pop(0)
    res.append(node.data)
    if node.left:
      queue.append(node.left)
    if node.right:
      queue.append(node.right)

  return res

if __name__ == '__main__':
  tree = Tree()
  tree.root = Node("A")
  tree.root.left = Node("B")
  tree.root.right = Node("C")
  tree.root.left.left = Node("D")
  tree.root.left.right = Node("E")
  tree.root.right.left = Node("F")
  tree.root.left.left.left = Node("G")

  print(preorder_list_recur(tree.root))
  print(inorder_list_recur(tree.root))
  print(postorder_list_recur(tree.root))
  print(lvlorder_list_stack(tree.root))
```

```python
# page81

def make_tree(arr):
  if not arr:
    return

  tree = Tree()
  tree.root = Node(arr[0])

  queue = [tree.root]
  idx = 1
  while queue and idx < len(arr):
    node = queue.pop(0)

    if idx < len(arr) and arr[idx] is not None:
      node.left = Node(arr[idx])
      queue.append(node.left)
    idx += 1
    if idx < len(arr) and arr[idx] is not None:
      node.right = Node(arr[idx])
      queue.append(node.right)
    idx += 1

  return tree

arr = ["A","B","C","D","E","F",None,"G"]
tree = make_tree(arr)

print(preorder_list_recur(tree.root))
print(inorder_list_recur(tree.root))
print(postorder_list_recur(tree.root))
print(lvlorder_list_stack(tree.root))
```

```python
# page93

def find_no_siblings(tgt):
  res = []

  queue = [tgt.root]
  while queue:
    node = queue.pop(0)
    if node.left is not None:
      queue.append(node.left)
      if node.right is None:
        res.append(node.left.data)

    if node.right is not None:
      queue.append(node.right)
      if node.left is None:
        res.append(node.right.data)

  return res if res else [-1]

tree = make_tree([37,20,None,113,None,None,None])
print("q=",lvlorder_list_stack(tree.root))
print(find_no_siblings(tree))
tree = make_tree([1,2,3])
print("q=",lvlorder_list_stack(tree.root))
print(find_no_siblings(tree))
tree = make_tree(['A','B','C','D','E','F',None,'G'])
print("q=",lvlorder_list_stack(tree.root))
print(find_no_siblings(tree))
```
