```python
# page 20~23

from bs4 import BeautifulSoup as bs
from pprint import pprint
import requests

headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3'
}

html = requests.get("https://www.nasa.gov/2025-news-releases/", headers=headers)
#pprint(html.text)

soup = bs(html.text, 'html.parser')
#pprint(soup)

# 나사 뉴스 영역 가져오기
data1 = soup.find('div', {'class':'hds-content-items'})
#print(type(data1))
#pprint(data1)

data2 = data1.find_all('div', {'class':'hds-content-item'})
#pprint(data2)

news_titles = []
for item in data2:
  item_area = item.find('div', {'class':'hds-content-item-inner'})
  #pprint(item_area)
  news_titles.append(item_area.find('div').text)

print(news_titles)
```

```python
# page 27~28

from bs4 import BeautifulSoup as bs
from pprint import pprint
import requests

headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3'
}

html = requests.get("https://www.nasa.gov/gallery/", headers=headers)
#pprint(html.text)

soup = bs(html.text, 'html.parser')
#pprint(soup)

image_div_list = soup.find_all('div', {'class':'hds-content-item'})

anchor_list = []

for img in image_div_list:
  # 이미지 영역 추출
  anchor_list.extend(img.find('a'))

#<img alt="Current Aeronautics Image Gallery" src="https://images-assets.nasa.gov/image/AFRC2025-0030-83/AFRC2025-0030-83~large.jpg?w=1920&amp;h=1279&amp;fit=clip&amp;crop=faces%2Cfocalpoint"/>, <img alt="Shuttle Testing Images" src="https://www.nasa.gov/wp-content/uploads/2025/05/ssc-19750508-s00011-ddd988.jpg?w=819"/>, <img alt="Astronaut Matthew Dominick’s Photos from Space" src="https://images-assets.nasa.gov/image/iss071e609065/iss071e609065~large.jpg?w=1920&amp;h=1280&amp;fit=clip&amp;crop=faces%2Cfocalpoint"/>, <img alt="Astronaut Don Pettit’s Photos from Space" src="https://www.nasa.gov/wp-content/uploads/2025/05/pettit-irosa.jpg?w=1024"/>, <img alt="NASA Spacewalkers" src="https://www.nasa.gov/wp-content/uploads/2025/05/54492396655-c6ba83df1d-o.jpg?w=1024"/>, <img alt="Cyclotron Image Gallery" src="https://www.nasa.gov/wp-content/uploads/2025/03/grc-1963-c-65331b.jpg?w=1024"/>, <img alt="2023 Iceland Field Campaign" src="https://www.nasa.gov/wp-content/themes/nasa/assets/images/404-bg.jpg"/>, <img alt="Special Projects Laboratory Image Gallery" src="https://www.nasa.gov/wp-content/uploads/2025/04/grc-1946-c-16086b.jpg?w=1024"/>, <img alt="From Italy to Arizona: HALO’s Journey" src="https://images-assets.nasa.gov/image/jsc2025e034730/jsc2025e034730~large.jpg?w=1919&amp;h=1279&amp;fit=clip&amp;crop=faces%2Cfocalpoint"/>, <img alt="Solar Electric Propulsion" src="https://images-assets.nasa.gov/image/GRC-2025-C-01680/GRC-2025-C-01680~large.jpg?w=1920&amp;h=1280&amp;fit=clip&amp;crop=faces%2Cfocalpoint"/>]
print(len(anchor_list), anchor_list)

for anchor in anchor_list:
  title = anchor['alt']
  img_src = anchor['src']
  print(title, img_src, end='\n')

```
