```python
# page 55 - 트위치 클립 다운로드
# ==> 트위치가 한국 서비스가 종료되어 클립 영상은 한국에서 접속할 수가 없습니다. 따라서 다른 video태그가 있는 사이트로 변경해서 진행하였습니다

from selenium import webdriver
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By
import sys
from selenium.webdriver.common.keys import Keys
import urllib.request
import os
from urllib.request import urlretrieve

import time
import pandas as pd
import chromedriver_autoinstaller  # setup chrome options
from datetime import date
import re

chrome_path = "/content/drive/MyDrive/Colab Notebooks/chromedriver"

sys.path.insert(0,chrome_path)
chrome_options = webdriver.ChromeOptions()

user_agent = "Mozilla/5.0 (Windows NT 10.0; WOW64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/136.0.0.0 Safari/537.36"
chrome_options.add_argument('user-agent=' + user_agent)
chrome_options.add_argument('--headless') # ensure GUI is off
chrome_options.add_argument('--no-sandbox')
chrome_options.add_argument('window-size=1920*1080')
chrome_options.add_argument('disable-gpu')
chrome_options.add_argument('--disable-dev-shm-usage')  # set path to chromedriver as per your configuration
chrome_options.add_argument('lang=ko_KR') # 한국어

chromedriver_autoinstaller.install()  # set the target URL

driver = webdriver.Chrome(options=chrome_options)
driver.get("https://mixkit.co/free-stock-video/close-up-of-sunlit-water-ripple-101192/")
driver.implicitly_wait(300) # element가 로드될 때까지 지정한 시간만큼 대기할 수 있도록 설정

url_element = driver.find_element(By.TAG_NAME, 'video')

# VIDEO URL
print(url_element.get_attribute('src'))
video_url = url_element.get_attribute('src')

# TITLE
title_element = driver.find_element(By.CLASS_NAME, 'video-item-page__title')

print(title_element.text)
video_title = re.sub('[^0-9a-zA-Zㄱ-힗]', '', title_element.text)

# DATE
print(date.today())
video_date = date.today()

urlretrieve(video_url, str(video_date)+"_"+video_title+"_.mp4")


driver.close()
```

```python
# page 66

import smtplib
from email.mime.multipart import MIMEMultipart
from email.mime.text import MIMEText
from email.mime.application import MIMEApplication

sendEmail = "mypsyche98@gmail.com"
recvEmail = "mypsyche98@gmail.com"
password = "uovi hhug mnob mzwh"

smtpName = "smtp.gmail.com"
smtpPort = 587

#여러 MIME을 넣기위한 MIME/Multipart
msg = MIMEMultipart()

msg["Subject"] = "코랩/파이썬 메일 전송 테스트 - 파일첨부"
msg["From"] = sendEmail
msg["To"] = recvEmail

# 본문추가
text = "이것은 코랩에서 파이선으로 작성된 smtp를 이용한 메일 전송 샘플입니다."
contentPart = MIMEText(text, _charset="utf8")
msg.attach(contentPart)

# 파일 추가
etcFileName = "2025-05-22_CloseUpofSunlitWaterRipple_.mp4"
with open(etcFileName, "rb") as etcFD:
  etcPart = MIMEApplication(etcFD.read())
  etcPart.add_header('Content-Disposition','attachment',filename=etcFileName)
  msg.attach(etcPart)

s = smtplib.SMTP(smtpName, smtpPort)
s.starttls()
s.login(sendEmail, password)
s.sendmail(sendEmail, recvEmail, msg.as_string())
s.close()
```
