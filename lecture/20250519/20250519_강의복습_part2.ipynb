```python
# page 43

import threading
import time

class Worker(threading.Thread):
  def __init__(self, name):
    super().__init__()
    self.name = name

  def run(self):
    print("sub thread start", threading.current_thread().name) # currentThread() is deprecated
    time.sleep(2)
    print("sub thread end", threading.current_thread().name) # getName() is deprecated


print("main thread start")
threads = []

for i in range(3):
    thread = Worker(i)
    thread.start()
    threads.append(thread)

for thread in threads:
  thread.join()

print("main thread post job")
print("main thread end")
"""
main thread start
sub thread start 0
sub thread start 1
sub thread start 2
sub thread end 0
sub thread end 1
sub thread end 2
main thread post job
main thread end
"""
```
