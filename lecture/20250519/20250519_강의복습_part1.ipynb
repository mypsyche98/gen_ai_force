```python
# page 12
class Car:
  def __init__(self, model):
    self.model = model

  @property
  def get_model(self):
    return self.model

c = Car("GV80")
print(c.get_model)
```

```python
# page 13
def outer(out1):
  def inner(in1):
    print("inner function called")
    print("outer argument:", out1)
    print("inner argument:", in1)
  return inner

f = outer(1)
f(10)
```

```python
# page 15
def outer(out1):
  def inner(in1):
    print("inner function called")
    print("inner argument:", in1)
    out1()
  return inner

def hello():
  print("hello")

f = outer(hello)
f(10)
```

```python
# page 17
def inner2():
  print("inner2 function is called")

inner2()
```

```python
# page 18,20
def deco2(f):
  def wrapper():
    print("-" * 20)
    f()
    print("-" * 20)
  return wrapper

decorated_inner = deco2(inner2)
decorated_inner()
```

```python
# page 21
@deco2
def inner3():
  print("inner function called")

inner3()
```
