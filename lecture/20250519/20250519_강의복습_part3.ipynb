```python
# page 52
import multiprocessing as mp
import time

def worker():
  proc = mp.current_process()
  print(proc.name)
  print(proc.pid)
  time.sleep(3)
  print("SubProcess End")

if __name__ == "__main__":
  #process spawning
  p = mp.Process(name="SubProcess", target=worker)
  p.start()

  time.sleep(4)
  #main process
  proc = mp.current_process()
  print(proc.name)
  print(proc.pid)

  print("MainProcess end")

""" 메인프로세스가 먼저 끝나는 경우 코랩출력캡처에 나오지 않기때문에 코드 순서를 변경하여 서브가 먼저끝나도록 수정해봄
SubProcess
1820
SubProcess End
MainProcess
1593
MainProcess end
"""
```
