```python
# 07-2 클로저와 데코레이터

# 클로저는 함수 안에 내부 함수를 구현하고 그 내부 함수를 리턴하는 함수

def mul3(n):
  return n*3

def mul5(n):
  return n*5

class Mul:
  def __init__(self, m):
    self.m = m

  def mul(self, n):
    return self.m * n

if __name__ == '__main__':
  mul3 = Mul(3)
  mul5 = Mul(5)

  print(mul3.mul(10))
  print(mul5.mul(10))
```

```python
class Mul:
  def __init__(self, m):
    self.m = m

  def __call__(self, n):
    return self.m * n

if __name__ == "__main__":
  mul3 = Mul(3)
  mul5 = Mul(5)

  print(mul3(10))
  print(mul5(10))
```

```python
def mul(m): # 이것이 클로저다!!!!
  def wrapper(n):
    return m * n
  return wrapper

if __name__ == "__main__":
  mul3 = mul(3)
  mul5 = mul(5)

  print(mul3(10))
  print(mul5(10))
```

```python
# 데코레이터 - 기존의 함수를 변경하지 않고 기능을 추가 할 수 있도록 만든 클로저를 데코레이터라고 한다
import time

def myfunc():
  start = time.time()
  print("함수가 실행됩니다.")
  end = time.time()
  print("running time %f sec" % (end-start))

myfunc()
```

```python
import time

def elapsed(org_func):
  def wrapper():
    start = time.time()
    result = org_func()
    end = time.time()
    print("elapsed time %f sec" % (end-start))
    return result
  return wrapper

def myfunc():
  print("myfunc is running")

decorated_myfunc = elapsed(myfunc)
decorated_myfunc()
```

```python

@elapsed
def myfunc2():
  print("myfunc2 is running")

myfunc2()
```

```python
@elapsed
def myfunc3(msg):
  print(f"msg={msg}")

myfunc3("test")
```

```python
def elapsed2(org_func):
  def wrapper(*args, **kwargs):
    start = time.time()
    result = org_func(*args, **kwargs)
    end = time.time()
    print("elapsed time %f sec" % (end-start))
    return result
  return wrapper

@elapsed2
def myfunc4(msg):
  print(f"msg={msg}")

myfunc4("test")
```
