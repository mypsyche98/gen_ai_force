```python
# 06-1 내가 프로그램을 만들 수 있을까?

def gugu(n):
  result = []
  i = 1
  while i < 10:
    result.append(n * i)
    i += 1
  return result

print(gugu(2))
```

```python
# 06-2 3과 5읠 배수를 모두 더하기

total = 0
for i in range(1,1000):
  if i % 3 == 0 or i % 5 == 0:
    total += i
print(total)
```

```python
# 06-3 게시판 페이징하기

def get_total_page(m, n):
  if m % n == 0:
    return m // n
  else:
    return m // n + 1

print(get_total_page(5,10))
print(get_total_page(15,10))
print(get_total_page(25,10))
print(get_total_page(30,10))
```
