```python
class Bird:
  def fly(self):
    raise NotImplementedError

class Eagle(Bird):
  pass

eagle = Eagle()
eagle.fly()
```

```python
class Eagle(Bird):
  def fly(self):
    print("Very fast")

eagle = Eagle()
eagle.fly()
```

```python
class MyError(Exception):
  def __str__(self):
    return 'NotAcceptableNick'

def say_nick(nick):
  if nick == 'fool':
    raise MyError()
  print(nick)

say_nick('angel')
say_nick('fool')
```

```python
try:
  say_nick('angel')
  say_nick('fool')
except MyError as e:
  print("not acceptable nick", e)
```

```python
def positive(l):
  result = []
  for i in l:
    if i>0 :
      result.append(i)
  return result

print(positive([1,-3,2,0,-5,6]))
```

```python
def positive(x):
  return x > 0

print(list(filter(positive, [1,-3,2,0,-5,6])))
```
