```python
coffee = 10
money = 300
while money:
  print("recv money, gave coffee")
  coffee = coffee - 1
  print("remain coffee %d" % coffee)
  if coffee == 0:
    print("soldout coffee")
    break
```

```python
coffee = 10

while True:
  money = int(input("insert money:"))
  if money == 300:
    print("serve coffee")
    coffee -= 1
  elif money > 300:
    print("charge %d, serve coffee" % (money-300))
    coffee -= 1
  else:
    print("refund money, do not serve coffee")
    print("remain coffee is %d" % coffee)

  if coffee == 0:
    print("soldout coffee")
    break
```

```python
marks = [90,25,67,45,80]

number = 0
for mark in marks:
  number = number + 1
  if mark >= 60:
    print("number %d is approved" % number)
  else:
    print("number %d is failed" % number)
```
