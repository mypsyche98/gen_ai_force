```python
# page 53 - exam 05

fruits = {"apple":"red", "banana":"yellow", "grape":"purple", "melon":"green"}

for color in fruits.values():
  print(color, end = ' ')
print()

for fruit in fruits.keys():
  print(fruit, end = ' ')
print()

for fruit, color in fruits.items():
  print("%s의 색은 %s" %(fruit, color), end = ', ')
```
