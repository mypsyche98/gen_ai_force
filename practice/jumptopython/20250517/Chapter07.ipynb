```python
# page 47 - exam 03

str = "Hello goorm!!!"
print(str, type(str))

s0 = set(str)
print(s0, type(s0))

newstr = tuple(s0)
print(newstr, newstr[4], newstr[5:], type(newstr))

l = [1,2,3,4,5,6,7,8]
print(l, type(l))

s1 = set(l)
print(s1, type(s1))

newlist = list(s1)
print(newlist, newlist[4], newlist[:5], type(newlist))
```

```python
# page 47 - exam 04

s1 = set([2,4,6,8,10])
s2 = set([1,2,3,4,5,6,7,8])

interset = s1 & s2
print(interset)
print(s1.intersection(s2), s2.intersection(s1))
print(s1)

uniset = s1 | s2
print(uniset)
print(s1.union(s2))
print(s1)

diffset1 = s1 - s2
diffset2 = s2 - s1
print(diffset1)
print(diffset2)
```
