```python
# page 58 - exam 02

wordslist = []

while True :
  word = input("단어를 입력하세요. 1 입력시 중단됩니다.")
  if word == '1':
    break
  else :
    wordslist.append(word)

for w in wordslist:
  print(w, end = ' ')
```

```python
# page 58 - exam 03

nums = []

while True :
  num = int(input("숫자를 입력하세요. -1 입력시 중단됩니다.:"))
  if num == -1:
    break
  else :
    if num % 2 == 1 :
      continue
    nums.append(num)

for n in nums:
  print(n, end = ' ')
```
