```python
from operator import itemgetter

students = [
    {"name":"jane", "age":22, "grade":'A'},
    {"name":"dave", "age":32, "grade":"B"},
    {"name":"sally", "age":17, "grade":"B"},
]

result = sorted(students, key=itemgetter('name'))
print(result)
```

```python
from operator import attrgetter

class Student:
  def __init__(self, name, age, grade):
    self.name = name
    self.age = age
    self.grade = grade

  def __str__(self):
    return f"name={self.name}/age={self.age}/grade={self.grade}"

students = [
    Student("jane", 22, "A"),
    Student("dave", 32, "B"),
    Student("sally", 17, "B"),
]

result = sorted(students, key=attrgetter('name'))
print(result)
```

```python
import time

def long_task():
  for i in range(5):
    time.sleep(0.3)
    print("working:%s\n" % i)

print("Start")

for i in range(5):
  long_task()

print("End")
```

```python
import time
import threading

def long_task():
  for i in range(5):
    time.sleep(0.5)
    print("working:%s" % i)

print("Start")

threads = []
for i in range(5):
  t = threading.Thread(target=long_task)
  threads.append(t)

for t in threads:
  t.start()

for t in threads:
  t.join()

print("End")
```

```python
import urllib.request

def get_wikidocs(page):
  resource = "https://wikidocs.net/{}".format(page)
  with urllib.request.urlopen(resource) as s:
    with open('wikidocs_%s.html' % page, 'wb') as f:
      f.write(s.read())

get_wikidocs(33)
```
