```python
# 리스트의 인덱싱과 슬라이싱

a = [1,2,3]
print(a)

print(a[0])

print(a[0]+a[2])

print(a[-1])

a = [1,2,3,['a','b','c']]

print(a[0])

print(a[-1])

print(a[3])

print(a[-1][0])

print(a[-1][-1])

a = [1,2,['a','b',['Life','is']]]

print(a)
print(a[-1][-1][0])
```

```python
a = [1,2,3,4,5]
print(a[0:2])

a = "12345"
print(a[0:2])

a = [1,2,3,4,5]
b = a[:2]
c = a[2:]

print(b)
print(c)

a = [1,2,3,['a','b','c'],4,5]
print(a[2:5])
print(a[3][:2])
```

```python
a = [1,2,3]
b = [4,5,6]

print(a+b)
print(a*3)

print(len(a))

a[2] = 4
print(a)

del a[1]
print(a)
```

```python
a = [1,2,3,4,5]

del a[2:]
print(a)
```

```python
a = [1,2,3]
a.append(4)
print(a)

a.append([5,6])
print(a)

a = [1,4,3,2]
a.sort()
print(a)
print(a.index(3))
print(a.index(1))

a.insert(0,4)
print(a)
a.insert(3,5)
print(a)

a = ['c','b','a']
a.sort()
print(a)

a.reverse()
print(a)

```

```python
a = [1,2,3,1,2,3]
a.remove(3)
print(a)
a.remove(3)
print(a)

```
