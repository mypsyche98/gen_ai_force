```python
class FourCal:
  def setdata(self, first, second):
    self.first = first
    self.second = second
  def add(self):
    result = self.first + self.second
    return result
  def mul(self):
    result = self.first * self.second
    return result
  def sub(self):
    result = self.first - self.second
    return result
  def div(self):
    result = self.first / self.second
    return result


a = FourCal()
b = FourCal()
a.setdata(4,2)
b.setdata(3,8)

print(f"a.add()={a.add()}")
print(f"a.mul()={a.mul()}")
print(f"a.sub()={a.sub()}")
print(f"a.div()={a.div()}")
print(f"b.add()={b.add()}")
print(f"b.mul()={b.mul()}")
print(f"b.sub()={b.sub()}")
print(f"b.div()={b.div()}")
```

```python
class FourCal:
  def __init__(self, first, second):
    self.first = first
    self.second = second
  def setdata(self, first, second):
    self.first = first
    self.second = second
  def add(self):
    result = self.first + self.second
    return result
  def mul(self):
    result = self.first * self.second
    return result
  def sub(self):
    result = self.first - self.second
    return result
  def div(self):
    result = self.first / self.second
    return result


a = FourCal(4,1)
print(a.add())
```

```python
class MoreFourCal(FourCal):
  def pow(self):
    result = self.first ** self.second
    return result

a = MoreFourCal(4,2)
print(f"a.add()={a.add()}")
print(f"a.mul()={a.mul()}")
print(f"a.sub()={a.sub()}")
print(f"a.div()={a.div()}")
print(f"a.pow()={a.pow()}")
```

```python
class SafeFourCal(FourCal):
  def div(self):
    if self.second == 0:
      return 0
    else:
      return self.first / self.second

  a = SafeFourCal(4,0)
  print(a.div())
  a.setdata(4,2)
  print(a.div())
```
