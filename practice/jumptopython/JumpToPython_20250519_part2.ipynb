```python
# 02-6 집합자료형

s1 = set([1,2,3])
print(s1)

s2 = set("hello")
print(s2)
```

```python
s1 = set([1,2,3])
l1 = list(s1)
print(l1)

print(l1[0])

t1 = tuple(s1)
print(t1)
print(t1[0])
```

```python
s1 = set([1,2,3,4,5,6])
s2 = set([4,5,6,7,8,9])

print(s1 & s2)
print(s1.intersection(s2))

print(s1 | s2)
print(s2.union(s2))

print(s1 - s2)
print(s2 - s1)
print(s1.difference(s2))
print(s2.difference(s1))

```

```python
s1 = set([1,2,3])
s1.add(4)
print(s1)

s1.update([4,5,6])
print(s1)

s1.remove(2)
print(s1)
```

```python
# 02-8 자료형의 값을 저장하는 공간, 변수

a = [1,2,3]
print(id(a))

b = a
print(a, b)

a[1] = 4
print(a, b)

print(a is b)
```

```python
a = [1,2,3]
b = a[:]
print(a, b, a is b)
a[1] = 4
print(a,b)
```

```python
from copy import copy
a = [1,2,3]
b = copy(a)
print(b is a)

c = b.copy()
print(c is b)
```
