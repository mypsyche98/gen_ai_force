```python
# page 69 - exam 01

nums = input("두 수를 띄어쓰기로 입력하세요:").split()
nums = list(map(int, nums))

result = 0
cal = input("""
1. +
2. -
3. *
4. /
""")

if cal == "1":
  result = nums[0] + nums[1]
elif cal == "2":
  result = nums[0] - nums[1]
elif cal == "3":
  result = nums[0] * nums[1]
elif cal == "4":
  result = nums[0] / nums[1]
else :
  print("잘못된 입력입니다.")
print(result)
```
