```python
# page 61 - exam 02

def stdInfo(rrn):
  if int(rrn[6]) == 3 or int(rrn[6]) == 4:
    biryear = int(rrn[0]+rrn[1]) + 2000
  else :
    biryear = int(rrn[0]+rrn[1]) + 1900

  age = year - biryear + 1
  birmonth = int(rrn[2] + rrn[3])
  birday = int(rrn[4] + rrn[5])

  if rrn[6] == "2" or rrn[6] == "4":
    gen = "female"
  else :
    gen = "male"

  return [age, birmonth, birday, gen]


year = 2025
stdlist = []

stdrrn = input("학생의 주민등록번호를 7자리 입력하세요.")

stdlist.append(stdInfo(stdrrn))

print(stdlist[0])
```

```python
# page 63 - exam 01

def guide():
  print("두 정수를 입력받아 곱한 결과를 출력하는 프로그램입니다.")

def inputnums():
  a = int(input("첫번째 정수를 입력하세요."))
  b = int(input("두번째 정수를 입력하세요."))
  return a, b

def multi(num1, num2) :
  return num1 * num2

def printResult(result):
  print("두 수의 곱셈 결과는 %d입니다." %result)

guide()
n1, n2 = inputnums()
res = multi(n1, n2)
printResult(res)
```

```python
# page 64 - exam 05

def calNums(ch, *t):
  if ch ==  "sum":
    total = 0
    for i in t:
      total += i
  elif ch == "mul":
    total = 1
    for i in t:
      total *= i
  else :
    print("실행할 수 없습니다")
  return total

choice = input("덧셈은sum, 곱셈은 mul을 입력하세요:")
print(calNums(choice,1,2,3,2,5,3,2))
```
