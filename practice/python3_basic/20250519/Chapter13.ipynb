```python
# page83 - exam 01
#file이름 : calculator.py

def add(a,b):
  return(a+b)

def sub(a,b):
  return(a-b)

def mul(a,b):
  return(a*b)

def div(a,b):
  return(a/b)

def mod(a,b):
  return(a%b)
```

```python
# page83 - exam 02
import calculator

add_result = calculator.add(10,2)
sub_result = calculator.sub(10,2)
mul_result = calculator.mul(10,2)
div_result = calculator.div(10,2)
mod_result = calculator.mod(10,2)

print(add_result)
print(sub_result)
print(mul_result)
print(div_result)
print(mod_result)
```
