```python
# 18-01 열 인덱스 반복자 반환 (__iter__)

import pandas as pd
import numpy as np

data = {'col1':[1,2],'col2':[3,4]}
df = pd.DataFrame(data = data)
print(df)
```

```python
df2 = df.__iter__()
print(df2)
print()
print(next(df2))
print()
print(next(df2))
print()
print(list(df))
```

```python
# 18-02 열과 내용의 반복자 반환 (items, iteritems)

data = {'col1':[1,2],'col2':[3,4]}
idx = ['row1','row2']
df = pd.DataFrame(data = data, index=idx)
print(df)
```

```python
df2 = df.items()
print(df2)
print()
df2 = df.items()
for i in df2:
    print("="*100)
    print(i)
print()
df2 = df.items()
for i in df2:
    print("="*30)
    for j in i:
        print(j)
```

```python
# 18-03 행과 내용의 반복자 반환 (iterrows)

data = {'col1':[1,2],'col2':[3,4]}
idx = ['row1','row2']
df = pd.DataFrame(data = data, index=idx)
print(df)
```

```python
df2 = df.iterrows()
print(df2)
print()
df2 = df.iterrows()
for i in df2:
    print("="*30)
    print(i)
print()
df2 = df.iterrows()
for i in df2:
    print("=" * 30)
    for j in i:
        print(j)
```

```python
# 18-04 튜플형태 반복자 반환 (itertuples)

data = {'col1':[1,2],'col2':[3,4]}
idx = ['row1','row2']
df = pd.DataFrame(data = data, index=idx)
print(df)
```

```python
print(df.itertuples())
print()
print(list(df.itertuples(index=False)))
print()
print(list(df.itertuples(name=None)))
print()
print(list(df.itertuples(name="테스트")))
```
